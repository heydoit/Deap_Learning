```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
import numpy as np
import pandas as pd
```

```python
red_df = pd.read_csv('./winequality-red.csv')
redwine_X = red_df.iloc[:,:-1]
redwine_y = red_df.iloc[:,-1]
redwine_y.value_counts()
redwine_y1 = redwine_y-3
redwine_y1.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(redwine_X, redwine_y1,
                                                   test_size=0.3)
```

```python
model = Sequential()
# 첫번째 h1
model.add(Dense(100, input_shape=(11,), activation='sigmoid')) # input_shape은 features
model.add(Dropout(0.2))

# 두번째 h2
model.add(Dense(200, activation='relu'))
model.add(Dropout(0.4))

# 세번째 h3
model.add(Dense(50, activation='tanh'))
model.add(Dropout(0.1))

# 출력층 정의
model.add(Dense(6, activation='softmax'))
```

```python
# 학습을 정의한다.

model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(X_train,y_train.to_numpy(),
         epochs = 300, batch_size=200)
```

```python
model.evaluate(X_test, y_test)
```
